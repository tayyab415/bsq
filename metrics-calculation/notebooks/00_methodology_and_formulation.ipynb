# BSQ methodology & formulation

**Body-Strike Quality (BSQ)** scores every shot from Bundesliga **3D skeleton** (50 Hz, 21 parts) fused with **KPI events**.

This notebook is the **research layer**: phases P1–P6, modules, and citations. Computation is in `01_pipeline_computation.ipynb`; league and player visuals in `02` and `03`.

Judges can read this repo **without AWS SSO** — reference outputs are bundled under `metrics-calculation/reference_outputs/` (derived scores only, no raw parquet).

## 1. Six phases (relative to contact frame)

| Phase | Frames (50 Hz) | Focus |
|-------|----------------|--------|
| P1 | −125 … −51 | Context: distance, angle, pressure, defenders, keeper |
| P2 | −50 … −14 | Approach & reception: speed, prep, timing |
| P3 | −13 … −2 | **Backswing / coil**: hip–shoulder separation, knee load, foot speed |
| P4 | −2 … +2 | **Impact**: contact ankle, plant foot, foot into ball |
| P5 | +2 … +25 | **Follow-through**: balance, COM continuation |
| P6 | +25 … +75 | **Outcome**: goal alignment, exit speed, launch |

Defined in `shooting1/metric.py` as `PHASE_OFFSETS`.

## 2. Module scores (what coaches see)

| Module | Built from | Example features |
|--------|------------|------------------|
| Decision quality | P1 + pass options | pressure, lane, xG audit |
| Shot geometry | P1 | distance, angle |
| Arrival / receiving | P2 | reception pressure |
| Approach / prep | P2 | approach speed, prep touches |
| **Technique mechanics** | P2–P5 | shoulder–hip, knee, contact, follow-through |
| **Strike quality** | P4 mech + strike | foot speed, plant, sequencing |
| **Placement** | P6 | goal alignment, KPI outcome |
| **Strike output** | P4 + P6 | exit speed, launch, trajectory |
| **Finishing execution index** | Post-strike cohort percentiles | strike quality, goal-plane lateral, exit speed, ball-to-foot ratio |

Full mapping: `MODULE_FEATURE_DECLARATIONS` in `metric.py`.

## 3. P3 backswing (coil) formulation

Hip–shoulder separation (X-factor) is computed in the **horizontal pitch plane** from skeleton pelvis and shoulder bearings (`_axis_separation` in `metric.py`). Peak separation in the P3 window feeds `shoulder_hip_score` alongside COM, knee stability, knee peak velocity, arm abduction, and striking-foot speed peak.

Family-specific weights (`P3_V3_COMPONENT_WEIGHTS`) up-weight COM for cutbacks, foot peak for volleys.
